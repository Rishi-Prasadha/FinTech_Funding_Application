```python
# Import the required packages 
import pandas as pd
from pathlib import Path
import hvplot.pandas
import numpy as np
```

```python
# Import the InsideAirbnb dataset into a dataframe

airbnb_df = pd.read_csv(Path("Resources/austin_listings.csv"))

# Preview the dataframe

display(airbnb_df.head())
```

```python
# Analyze the amount of property types available 

display(airbnb_df['room_type'].value_counts())

airbnb_df['room_type'].value_counts().hvplot.bar(title='Distribution of Airbnb Types in Austin Area')
```

```python
# Describe the Airbnb data for on Entire home/apt listings

entire_home_desc = airbnb_df['price'].loc[airbnb_df['room_type'] == 'Entire home/apt'].describe()
entire_home_desc = entire_home_desc.T
entire_home_desc
```

```python
entire_home_df = airbnb_df.loc[airbnb_df['room_type'] == 'Entire home/apt']
entire_home_df = entire_home_df.drop(columns=['neighbourhood_group','host_name','host_id','name','id','license'])
entire_home_df
```

```python
entire_home_df.hvplot.points(
    'longitude',
    'latitude',
    geo=True,
    color='price',
    tiles="OSM",
    size='neighborhood',
    ylabel='',
    xlabel='',
    title='Airbnb Entire Home/Apt in the Austin Area'
)
```

```python
avgprice_by_zip = entire_home_df.groupby('neighbourhood')['price'].mean()
avgprice_by_zip
```
