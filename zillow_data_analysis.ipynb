```python
# Import the required packages 
import pandas as pd
from pathlib import Path
import hvplot.pandas
import numpy as np
```

```python
# Import the Zillow Home Value dataset into a dataframe

zillow_df = pd.read_csv(
    Path("Resources/zillow_home_values.csv"),
    infer_datetime_format=True
)
# Preview the dataframe

display(zillow_df.head())
```

```python
zillow_austin_df = zillow_df.loc[zillow_df['City'] == 'Austin']
zillow_austin_tx_df = zillow_austin_df.loc[zillow_austin_df['State'] == 'TX']
display(zillow_austin_tx_df.head())
```

```python
# Create a dataframe with only the prices and dates and another dataframe with zip codes for headers

dates_df = zillow_austin_tx_df.drop(columns=['RegionID', 'RegionName','RegionType','State', 'City', 'Metro', 'CountyName', 'SizeRank','StateName']).T
zipcode_df = zillow_austin_tx_df['RegionName']



dates_df.set_index(dates_df.columns[0])
dates_df.columns = zipcode_df

display(dates_df)
```

```python
dates_df.hvplot.line(rot=90, 
    width=2500, 
    height=600
).opts(yformatter="%.0f")
```

```python
# Iterate through the dataframe and fit all zip codes to an exponential regression model


#display(len(days_elapsed))
#display(len(dates_df[78704]))

#zip1_fit = np.polyfit(days_elapsed, np.log(dates_df[78704]), 1)
#display(zip1_fit)

days_elapsed = [num for num in np.arange(0, 8100, 30)]
zip_fits = []
for (column_name, column_data) in dates_df.iteritems():
    try:
        current_fit = np.polyfit(days_elapsed, np.log(dates_df[column_name]), 1)
        zip_fits.append(current_fit)
    except: 
        zip_fits.append(np.nan)
    finally: 
        continue
    # try to drop the zip code with the nan


#pd.DataFrame(zip_fits)
#display(zip_fits)
#display(len(zip_fits))
```

```python
# Create two data frames to store the "slope" (m) and the 'intercept' (b) of each fit for each zip code
# ln(y) = mx + b
# y = b * m^x
b = []
m = []

for arr in zip_fits:
    try:
        m.append(arr.tolist()[0])
        b.append(arr.tolist()[1])
        #m.append(proxy_m)
        #b.append(proxy_b)
        #print(m)
        #print(b)
    except: 
        m.append(np.nan)
        b.append(np.nan)
    finally: 
        continue

#m = pd.Series(m)
#m.isna()
```

```python
# Merge zip codes and models into one dataframe

zip_fits_df = pd.DataFrame({
    'Zip Code': zipcode_df,
    'Slope': m,
    'Intercept': b
})

zip_fits_df.dropna(inplace=True)

display(len(zip_fits_df))
display(zip_fits_df)
```
